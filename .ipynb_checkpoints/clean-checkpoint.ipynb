```python
import pandas as pd

df = pd.read_csv("视频会员订单数据源.csv")
```

```python
df["price"]=df["price"]/100
df["pay_time"]=pd.to_datetime(df["pay_time"])
df["create_time"]=pd.to_datetime(df["create_time"])
```

```python
df.info()
```

```python
Platform = df[df["platform"].isnull()]
df.drop(index=Platform.index,inplace=True)
```

```python
Payment_provider = df[df["payment_provider"].isnull()]
df.drop(index=Payment_provider.index,inplace=True)
```

```python
df.info()
```

```python
result1 = df[~df["platform"].isin(["ios","android"])]
print(result1)
```

```python
result2 = df[~df["payment_provider"].isin(["applepay","wxpay","alipay"])]
print(result2)
```

```python
result3 = df[df["pay_time"]<df["create_time"]]
print(result3)
```

```python
df.drop(index=result3.index,inplace=True)
```

```python
df.info()
```

```python
result4 = df[df["order_id"].duplicated()]
print(result4)
```

```python
df.drop(index=result4.index,inplace=True)
```
